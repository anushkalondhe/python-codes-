```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x=np.array([25,29,33,37,44,47,57,59,60,67])
y=np.array([11,12,23,34,45,46,56,67,78,89])
z=np.array([87,82,78,73,67,63,55,51,47,39])
```

# line plot

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='*',mfc='lavender',color='purple',ls=':',ms=10);
plt.grid()

#markers= >,<,^,H,o,*
```

```python
plt.plot(y,z);
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='*',mfc='lavender',color='purple',ls=':',ms=10,label='2023');
plt.plot(x,z,marker='*',mfc='yellow',color='purple',ls=':',ms=10,label='2024');
plt.grid()
plt.legend();
```

# sub plot

```python

plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='*',mfc='lavender',color='purple',ls=':',ms=10);
plt.grid()

plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='*',mfc='yellow',color='purple',ls=':',ms=10,label='2024');
plt.grid()
```

# bar plot

```python
courses=['java','python','kotlin','dart','R']
students=[67,63,59,55,47]
plt.title('student enrollment')
plt.xlabel('courses')
plt.ylabel('students')
plt.bar(courses,students,color='lavender',edgecolor='black',width=0.5,linewidth=2,alpha=0.9);
plt.grid() 
```
